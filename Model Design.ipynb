```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Upsample_Churn.csv')
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['CUS_Gender'] = le.fit_transform(df['CUS_Gender'])
df['CUS_Marital_Status'] = le.fit_transform(df['CUS_Marital_Status'])
```

```python
df
```

```python
dict_new = {'PLATINUM':0, 'EXECUTIVE':1, 'MIDLE': 2, 'LOW': 3}
df.replace({'TAR_Desc': dict_new}, inplace=True)
df['TAR_Desc'].unique()
```

```python
df.info()
```

```python
x = df.drop(['Status', 'CUS_Customer_Since', 'Suggestions'], axis=1)
x
```

```python
y = df['Status']
y
```

```python
#Train Test Split

#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
def model_deploy(x,y,model):
    from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
    model_test = model
    model_test.fit(x_train, y_train)
    y_pred = model_test.predict(x_test)
    print(f"Precision Score = {precision_score(y_test, y_pred)}")
    print(f"Recall Score = {recall_score(y_test, y_pred)}")
    print(f"F1 Score = {f1_score(y_test, y_pred)}")
    print(f"Accuracy Score = {accuracy_score(y_test, y_pred)}")
    print(f"Score: {model_test.score(x_test,y_test)}\n")
    print(f"Classification Report:\n {classification_report(y_test, y_pred, labels=[0,1])}")
    return model_test
```

**Decision Tree Classification**

```python
model = DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
model_dt = model_deploy(x,y, model)
```

**Random Forest Classification**

```python
from sklearn.ensemble import RandomForestClassifier
model =RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
model_rf = model_deploy(x,y, model)
```
